```python
import fitz
import pycorrector
import re 
from zhon import hanzi
import string

def cn_cut_sentence(text):
    text = re.sub('\n','',text)
    sentences = re.split(r'(\!|\?|。|！|？|\.{6})',text)
    sentences = [sent for sent in sentences if sent not in (set(hanzi.punctuation)|set(string.punctuation))]
    sentences = [sent.strip() for sent in sentences ]
    return sentences


pycorrector.enable_char_error(enable=False)
pycorrector.set_custom_confusion_dict(path='./my_custom_confusion.txt')

fname = '2019年第一期都江堰兴堰投资有限公司停车场专项债券募集说明书.pdf'  # get document filename

out= open(fname[:4] + "文本纠错（添加白名单）.txt", "w")

red = (1, 0, 0)
blue = (0, 0, 1)
gold = (1, 1, 0)
green = (0, 1, 0)
def print_descr(annot, pos, correct_w):
    """Print a short description to the right of each annot rect."""
    annot.parent.insertText(
        annot.rect.br + ((-pos)*5, -20), "%s" % correct_w, fontfile = 'SimHei/SimHei.ttf',fontname ='FO',color= blue)

doc = fitz.open(fname)  # open document
for page in doc:  # iterate the document pages
    text = re.findall('\n \n\d+ \n(.*)',page.getText(), re.M|re.S)
    if text:
        sentences = cn_cut_sentence(text[0])
        for sent in sentences:
            correct_sent, detail = pycorrector.correct(sent)
            if detail:
                print('#########')
                print(f"页数{page.number}\n 修正后的句子** {correct_sent}\n ,错别字及修正后的字{detail}")
                out.write(str(page.number)+'\n')
                out.write(str(correct_sent)+'\n')
                out.write(str(detail)+'\n')
                for wrong in detail:
                    for txt in page.searchFor(wrong[0]):
                        annot = page.addHighlightAnnot(txt)
                    print_descr(annot,0,wrong[1])
doc.save(fname[:4] + "文本纠错（添加白名单）.pdf")

out.close()
```

```python
import fitz
import re

red = (1, 0, 0)
blue = (0, 0, 1)
gold = (1, 1, 0)
green = (0, 1, 0)

out= open(fname[:4] + "标点符号纠错.txt", "w")
doc = fitz.open(fname)

def print_descr(annot, pos, correct_w):
    """Print a short description to the right of each annot rect."""
    annot.parent.insertText(
        annot.rect.br + ((-pos)*5, -20), "%s" % correct_w, fontfile = 'SimHei/SimHei.ttf',fontname ='FO',color= blue)

patterns = {"[\u4E00-\u9FA5],":"，", "[\u4E00-\u9FA5]\?":"？", "[\u4E00-\u9FA5]!":"！", "[\u4E00-\u9FA5]:":"：",
            "[\u4E00-\u9FA5]\)":"）", "\([\u4E00-\u9FA5]":"（" }
for page in doc: 
    text = re.findall("\n \n\d+ \n(.*)",page.getText(), re.M|re.S)
    if text:
        sentences = cn_cut_sentence(text[0])
        for sent in sentences:
            for pattern in patterns:
                if re.findall(pattern, sent):
                    print(f"页数{page.number}\n ########")
                    out.write(f"页数{page.number}\n")
                    print(f"原句*** {sent}")
                    out.write(f"原句*** {sent}\n")
                    for wrong in re.findall(pattern, sent):
                        if pattern !=  "\([\u4E00-\u9FA5]":
                            print(f"wrong {wrong} right {wrong[0]}{patterns[pattern]}")
                            out.write(f"wrong {wrong} right {wrong[0]}{patterns[pattern]}\n")
                            for txt in page.searchFor(wrong):
                                annot = page.addHighlightAnnot(txt)                           
                        else:
                            print(f"wrong {wrong} right {patterns[pattern]}{wrong[1]}")
                            out.write(f"wrong {wrong} right {patterns[pattern]}{wrong[1]}\n")
                            for txt in page.searchFor(wrong):
                                annot = page.addHighlightAnnot(txt)                              
                        print_descr(annot, 1, patterns[pattern]) 
doc.save(fname[:4] +'标点符号纠错.pdf')
out.close()
```
